```python
import csv, sqlite3 # enables csv and sqlite3

con = sqlite3.connect("socioeconomic.db")
#connects sqlite to the database named socioeconomic.db
cur = con.cursor() #cursor is connected to the connection object
```

```python
%sql sqlite:///socioeconomic.db
#lets you ask sql queries
```

```python
import pandas
df = pandas.read_csv('jcxq-k9xf.csv')
df.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False,method="multi")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['per_capita_income_'], df['hardship_index'], color='blue')
plt.title('Scatter Plot of Per Capita Income vs Hardship Index')
plt.xlabel('Per Capita Income')
plt.ylabel('Hardship Index')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Per Capita Income vs Working population')
plt.scatter(df['per_capita_income_'],df['percent_aged_under_18_or_over_64'],color='red')
plt.xlabel('Per Capita Income')
plt.ylabel('Working Population')
plt.grid(True)
plt.show()
```
